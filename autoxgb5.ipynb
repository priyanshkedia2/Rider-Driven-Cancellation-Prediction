```python
from autoxgb import AutoXGB
```

```python
train_filename="./train.csv"
output="submission2"
test_filename="./test.csv"
idx="Id"
targets=["cancelled"]
use_gpu=True
num_folds=7
seed=42
num_trials=150
time_limit=500
fast = False
```

```python
axgb=AutoXGB(
    train_filename=train_filename,
    output=output,
    test_filename=test_filename,
    idx=idx,
    targets=targets,
    use_gpu=use_gpu,
    num_folds=num_folds,
    num_trials=num_trials,
    time_limit=time_limit
)
```

```python
axgb.train()
```
